```python
import pandas as pd
import json
import numpy as np
```

```python
second_vote_list = []
with open('vote.json', 'r') as f:
    data_raw = json.load(f)
data_raw
```

```python
results = data_raw['results']['Wahlkreise']
results
```

```python
for kreis, result in results.items():
    second_vote_result = result[0]
    second_vote_list.append([kreis, second_vote_result[0] + second_vote_result[1], second_vote_result[2], second_vote_result[3], second_vote_result[4], second_vote_result[5], second_vote_result[6], second_vote_result[7]])

second_vote_list
```

```python
x = 'Zollernalb – Sigmaringen (Wahlkreis 295)'
x.split('(')
```

```python
with open('vote.txt', 'r') as f:
    text = f.read()
    lines = text.split('\n')
    for index, i in enumerate(second_vote_list):
        kreis = i[0]
        for l in lines:
            res = l.split('\t')
            if kreis == res[1]:
                second_vote_list[index][0] = (res[2].split(' ('))[0]
                break
second_vote_list
```

```python
with open('vote.csv', 'w') as f:
    f.write('name,union,gruene,spd,linke,afd,fdp,bsw\n')
    for i in second_vote_list:
        f.write(f'{i[0]},{i[1]},{i[2]},{i[3]},{i[4]},{i[5]},{i[6]},{i[7]}\n')
    f.flush()
```

```python
df = pd.read_csv('vote.csv', index_col='name')
df
```

```python
df.describe()
```

```python
from sklearn.decomposition import PCA
```

```python
model = PCA(n_components=2)
df_pca = model.fit_transform(df)
df_pca
```

```python
import matplotlib.pyplot as plt
df_pca_df = pd.DataFrame(df_pca, index=df.index)
fig = plt.figure(1, (30, 30))
ax = fig.add_subplot(111)
labels = df.index.to_list()
max_party_names = df.idxmax()
print(max_party_names)
ax.scatter(df_pca[:, 0], df_pca[:, 1], s=20)
for index, i in enumerate(labels):
    ax.annotate(text=i, xy=(df_pca[index, 0], df_pca[index, 1]), c='red' if i in max_party_names.iloc[0] else 'black')
for party, kreis in max_party_names.to_dict().items():
    ax.scatter(df_pca_df.loc[kreis][0], df_pca_df.loc[kreis][1], c='yellow', s=150)
    ax.annotate(text=f'{party} MAX', xy=(df_pca_df.loc[kreis][0], df_pca_df.loc[kreis][1]), c='red', fontsize=28)
```
